```python
import fcsparser

path = "/Volumes/score/CytometryFiles/CelSeq_IndexFiles/G000287_Donoghue/Donoghue24May23/24May23_INX_NKC 079_LCE660.fcs"
meta, data = fcsparser.parse(path, reformat_meta=True)
meta
```

```python
metalist = list(meta.keys())
locnames = [item for item in metalist if item.startswith('INDEX SORTING LOCATIONS')]
locnames = sorted(locnames, key=lambda s: int(s.split('_')[1]))

wells = []
for locname in locnames:
    loclist = meta[locname]
    locs = loclist.split(";")

    for loc in locs:
        if loc:
            x, y = loc.split(",")
            new_x = chr(int(x) + ord('A'))
            new_y = str(int(y) + 1)
            wells.append(new_x + new_y)
```

```python
data = data.sort_values("Time")
data['well'] = wells
data
```
